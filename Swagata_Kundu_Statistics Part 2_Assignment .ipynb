# Theory Questions:

### 1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions based on data. It involves making an initial assumption (null hypothesis), collecting evidence (data), and determining whether the evidence is strong enough to reject the null hypothesis in favor of an alternative hypothesis.

###  2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- **Null Hypothesis (H₀):** A statement that there is no effect or no difference. It is the default or starting assumption.
- **Alternative Hypothesis (H₁ or Ha):** A statement that contradicts the null hypothesis, suggesting a real effect or difference.

For example:
- H₀: The mean of a population is equal to 50
- H₁: The mean of a population is not equal to 50

###  3. What is the significance level in hypothesis testing, and why is it important?
The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05, 0.01. It sets the threshold for statistical significance.

###  4. What does a P-value represent in hypothesis testing?
The **P-value** is the probability of obtaining a result at least as extreme as the one observed, assuming the null hypothesis is true.

###  5. How do you interpret the P-value in hypothesis testing?
- If **P-value ≤ α**, reject the null hypothesis (evidence supports the alternative hypothesis).
- If **P-value > α**, do not reject the null hypothesis (insufficient evidence against H₀).

### 6. What are Type I and Type II errors in hypothesis testing?
- **Type I Error (False Positive):** Rejecting H₀ when it is true (controlled by α).
- **Type II Error (False Negative):** Not rejecting H₀ when H₁ is true (controlled by β).

### 7. What is the difference between a one-tailed and a two-tailed test?
- **One-tailed test:** Tests for an effect in one direction (e.g., H₁: μ > μ₀).
- **Two-tailed test:** Tests for an effect in either direction (e.g., H₁: μ ≠ μ₀).

### 8. What is the Z-test, and when is it used?
The **Z-test** is used to test hypotheses about population means when the population variance is known and the sample size is large (n > 30).

### 10. What is the T-distribution, and when should it be used?
The **T-distribution** is used instead of the normal distribution when the sample size is small (n ≤ 30) and the population standard deviation is unknown.

### 11. What is the difference between a Z-test and a T-test?
- **Z-test:** Used when population standard deviation is known and sample size is large.
- **T-test:** Used when population standard deviation is unknown and/or sample size is small.

### 12. What is the T-test, and how is it used?
The **T-test** compares sample means to determine if there is a significant difference. Variants include:
- One-sample T-test
- Two-sample T-test
- Paired T-test

### 13. What is the relationship between Z-test and T-test?
Both are used to test means, but the T-test is more reliable for small samples. As the sample size increases, the T-distribution approaches the normal distribution used in Z-tests.

### 14. What is a confidence interval?
A **confidence interval (CI)** is a range of values that is likely to contain a population parameter with a certain level of confidence (e.g., 95%).

### 15. What is the margin of error?
The **margin of error** quantifies the uncertainty in the estimate. It affects the width of the confidence interval:

CI = estimate ± margin of error

### 16. How is Bayes' Theorem used in statistics?
**Bayes' Theorem** updates the probability of a hypothesis based on new evidence:

P(A|B) = [P(B|A) × P(A)] / P(B)

It’s used in Bayesian inference, where prior beliefs are updated with data.

### 17. What is the Chi-square distribution?
The **Chi-square distribution** is used in tests of independence and goodness of fit. It is based on squared deviations from expected values.

### 18. What is the Chi-square goodness of fit test?
This test checks whether an observed distribution matches an expected distribution. It is used with categorical data.

χ² = Σ((O - E)² / E)

### 19. What is the F-distribution?
The **F-distribution** is used to compare two variances. It is skewed and used in ANOVA and regression analysis.

### 20. What is an ANOVA test, and what are its assumptions?
**ANOVA (Analysis of Variance)** compares the means of three or more groups. Assumptions:
- Independence of observations
- Normally distributed groups
- Equal variances (homogeneity of variance)

### 21. What are the different types of ANOVA tests?
- **One-way ANOVA:** One independent variable.
- **Two-way ANOVA:** Two independent variables.
- **Repeated Measures ANOVA:** Same subjects used in all conditions.

### 22. What is the F-test, and how does it relate to hypothesis testing?
The **F-test** compares the variances of two populations or tests the overall significance in ANOVA. It uses the F-distribution.

### 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy.stats import norm

# Step 1: Define your sample data
sample_data = np.array([68, 72, 65, 70, 75, 71, 69, 67, 74, 66])

# Step 2: Define population parameters
population_mean = 70   # Known population mean
population_std = 5     # Known population standard deviation

# Step 3: Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 4: Compute the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Compute the p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 6: Set significance level
alpha = 0.05

# Step 7: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Step 8: Interpret results
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

### 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
import numpy as np
from scipy import stats

def one_sample_z_test(data, pop_mean):
    n = len(data)
    mean_sample = np.mean(data)
    std_sample = np.std(data, ddof=1)
    z = (mean_sample - pop_mean) / (std_sample / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

data = np.random.normal(loc=102, scale=8, size=30)
z_stat, p_val = one_sample_z_test(data, 100)
print(f"Z = {z_stat:.2f}, P = {p_val:.4f}")
```

### 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Known population parameters
population_mean = 100
population_std = 15

# Sample data
sample = np.array([102, 98, 101, 105, 100, 96, 97, 103, 99, 104])
sample_size = len(sample)
sample_mean = np.mean(sample)

# Significance level
alpha = 0.05

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed critical values
z_critical = stats.norm.ppf(1 - alpha / 2)

# P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"Critical Z-value (two-tailed): ±{z_critical:.2f}")
print(f"P-value: {p_value:.4f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# ----------------- Plotting ----------------- #
# Z-distribution curve
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Shading the rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (right)')

# Mark the calculated Z-score
plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')

# Reference lines
plt.axvline(-z_critical, color='black', linestyle='--', label=f'Critical Value = -{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle='--', label=f'Critical Value = +{z_critical:.2f}')

plt.title('Two-Tailed Z-Test and Decision Region')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

### 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    se = sigma / np.sqrt(n)
    z_alpha = stats.norm.ppf(1 - alpha)
    crit_val = mu0 + z_alpha * se

    y0 = stats.norm.pdf(x, mu0, se)
    y1 = stats.norm.pdf(x, mu1, se)

    plt.plot(x, y0, label="H0", color="blue")
    plt.plot(x, y1, label="H1", color="green")
    plt.fill_between(x, y0, where=(x > crit_val), color='red', alpha=0.4, label="Type I Error")
    plt.fill_between(x, y1, where=(x < crit_val), color='orange', alpha=0.4, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Error Visualization")
    plt.show()

plot_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=30)
```

### 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    h = stderr * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - h, mean + h

data = np.random.normal(100, 15, 50)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

### 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
def margin_of_error(data, confidence=0.95):
    n = len(data)
    stderr = stats.sem(data)
    z = stats.t.ppf((1 + confidence) / 2, n - 1)
    return z * stderr

data = np.random.normal(100, 20, 60)
moe = margin_of_error(data)
print(f"Margin of Error: {moe:.2f}")
```

### 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python



import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([40, 35, 25])

# Define expected proportions (uniform distribution assumed here)
expected_proportions = np.array([1/3, 1/3, 1/3])

# Scale expected frequencies to match total of observed frequencies
expected = expected_proportions * np.sum(observed)

# Perform Chi-square Goodness-of-Fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected.round(2))
print(f"\nChi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```

### 16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample2 = np.random.normal(loc=52, scale=5, size=25)   # Mean=52, Std=5

# Sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Determine F-ratio: larger variance in numerator
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Significance level
alpha = 0.05

# Critical values
f_critical_low = f.ppf(alpha / 2, dfn, dfd)
f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

# p-value
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Output results
print("Sample 1 Variance:", var1)
print("Sample 2 Variance:", var2)
print(f"F-statistic: {F:.4f}")
print(f"Degrees of freedom: ({dfn}, {dfd})")
print(f"Critical F-values: lower = {f_critical_low:.4f}, upper = {f_critical_high:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
if F < f_critical_low or F > f_critical_high:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

### 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check assumptions of ANOVA: normality, homogeneity of variances, and independence (visually).
    
    Parameters:
    groups (list of arrays): A list containing sample arrays from different groups.
    """
    print("=== Checking ANOVA Assumptions ===\n")
    
    # Normality test (Shapiro-Wilk) for each group
    print("1. Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        print(f"  Group {i+1}: W={stat:.4f}, p={p:.4f} ->", 
              "Normal" if p > 0.05 else "Not Normal")
    print("\n")
    
    # Homogeneity of variances (Levene’s Test)
    print("2. Homogeneity of Variances (Levene's Test):")
    stat, p = levene(*groups)
    print(f"  Levene statistic={stat:.4f}, p={p:.4f} ->", 
          "Equal variances" if p > 0.05 else "Unequal variances")
    print("\n")
    
    # Visual inspection for independence via boxplot and residual plot
    print("3. Visual Check for Independence (Box Plot & Residuals):")
    
    # Combine data
    data = np.concatenate(groups)
    labels = []
    for i, group in enumerate(groups):
        labels.extend([f'Group {i+1}'] * len(group))
    
    # Boxplot
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=labels, y=data, palette='Set2')
    plt.title("Boxplot of Groups (Visual Check)")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.tight_layout()
    plt.show()

    # Residual plot
    from scipy.stats import f_oneway
    f_stat, p_val = f_oneway(*groups)
    grand_mean = np.mean(data)
    residuals = [val - grand_mean for val in data]

    plt.figure(figsize=(10, 4))
    plt.scatter(range(len(residuals)), residuals, alpha=0.7)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.title("Residual Plot")
    plt.xlabel("Observation Index")
    plt.ylabel("Residual")
    plt.tight_layout()
    plt.show()
```

```python
np.random.seed(42)
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(55, 5, 30)
group3 = np.random.normal(60, 5, 30)

# Check assumptions
check_anova_assumptions([group1, group2, group3])
```

### 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

# Simulate data (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=100, scale=10, size=30)  # Population mean = 100, std dev = 10

# Sample statistics
n = len(data)
sample_variance = np.var(data, ddof=1)

# Hypothesized population variance (σ₀²)
sigma_0_squared = 100  # e.g., std deviation = 10

# Chi-square test statistic
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

# Degrees of freedom
df = n - 1

# Critical values for two-tailed test
alpha = 0.05
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Print results
print("=== Hypothesis Test for Population Variance ===")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"Chi-square Critical Values (α={alpha}): {chi2_lower:.4f}, {chi2_upper:.4f}")

# Decision
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: The population variance is significantly different from the hypothesized value.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference in variance.")
```

### 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

# Simulate data
# Group 1: 70 successes out of 100 trials
# Group 2: 60 successes out of 100 trials
successes1 = 70
n1 = 100

successes2 = 60
n2 = 100

# Calculate sample proportions
p1 = successes1 / n1
p2 = successes2 / n2

# Pooled proportion
P = (successes1 + successes2) / (n1 + n2)

# Calculate the standard error
SE = np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Z-statistic
Z = (p1 - p2) / SE

# P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

# Print results
print("=== Z-Test for Proportions ===")
print(f"Sample Proportions: p1 = {p1:.4f}, p2 = {p2:.4f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: The proportions are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference in proportions.")
```

### 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate two datasets with different variances
np.random.seed(42)

# Dataset 1 (low variance)
data1 = np.random.normal(loc=50, scale=5, size=50)

# Dataset 2 (high variance)
data2 = np.random.normal(loc=50, scale=10, size=50)

# Calculate sample variances
s1_squared = np.var(data1, ddof=1)
s2_squared = np.var(data2, ddof=1)

# F-statistic
F_stat = s1_squared / s2_squared

# Degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Critical value for two-tailed test at alpha = 0.05
alpha = 0.05
F_critical_upper = stats.f.ppf(1 - alpha / 2, df1, df2)
F_critical_lower = stats.f.ppf(alpha / 2, df1, df2)

# Print results
print("=== F-Test for Comparing Variances ===")
print(f"Sample Variances: s1² = {s1_squared:.4f}, s2² = {s2_squared:.4f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"Critical values (α={alpha}): F_lower = {F_critical_lower:.4f}, F_upper = {F_critical_upper:.4f}")

# Decision rule
if F_stat < F_critical_lower or F_stat > F_critical_upper:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: The variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference in variances.")

# Visualization (F-distribution and rejection region)
x = np.linspace(0, 5, 500)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
plt.fill_between(x, y, where=(x >= F_critical_upper), color='red', alpha=0.3, label=f'Rejection region (F > {F_critical_upper:.2f})')
plt.fill_between(x, y, where=(x <= F_critical_lower), color='red', alpha=0.3, label=f'Rejection region (F < {F_critical_lower:.2f})')

plt.axvline(F_stat, color='green', linestyle='--', label=f'F-statistic = {F_stat:.4f}')
plt.title('F-Test: Comparing Variances')
plt.xlabel('F-statistic')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
